# Python Data types

```python
# Python does have 4 basic data types
# - Int
# - Float
# - Str
# - Bool


x = 5
print(x)
print(type(x))

y = 3.55
Print(y)
print(type(y))

a = "Islam"
print(a)
print(type(a))

b = True # False
print(b)
print(type(b))
```

# Arithmetic Operations

```python
# Arithmetic Operations
#  + , - , * , / , % (Modulus / Remainder) , ** (Exponent)

x = 7
y = 5

print(x+y)
print(x-y)
print(x/y)
print(x*y)
print(x%y)
print(x**y)
```

# Comparison Operations

```python
# Comparison operation usually are used with If Else or While statement
# > , < , >= , <= , == , !=

x = 7
y = 4

print(x>y)
print(x<y)
print(x>=y)
print(x<=y)
print(x==y)
print(x!=y)
```

# Logical Operations

```python
# and,  or, not, in

x = 9
y = 3

print((x>5) and (y>5))
print((x>5) or (y>5))
print(not(x>y))

txt = "Hello world"
print('Hello' in txt)
```
