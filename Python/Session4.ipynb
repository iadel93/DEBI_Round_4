# While loop

```python
# Syntax 
#  while (Condition):
#        # code to be executed  
#  as long as the condition is True the code inside the while loop will keep executing.
#  NOTE : Be careful to avoid infinite loops!
#  You must break the condition from inside the loop or change the condition variable somewhere in the loop.

x = 0
while x < 5:
    print("Hello")
    x += 1
```

create a code to print out the multilplication of 5 from 5\*1 to 5\*10

```python
i = 1
while i <= 10:
    print(i*5)
    i += 1
```
