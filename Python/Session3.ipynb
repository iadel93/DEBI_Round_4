# If, Else

```python
# Syntax:
# if (Condition_1):
#   Statemnet_1      # Statment 1 is only executed if Condition_1 is True
# elif (Condition_2):
#   Statment_2      # Statement2 is only executed if Condition_1 is False and Condition_2 is True
# else:
#   Statement_3     # Statement_3 is only executed if all conditions are False

x = 0

if x > 0:
    print("X is positive")
elif x < 0:
    print("X is negative")
else:
    print("X is zero")
```

# For Loop

```python
# Syntax
# for {iterator} in {iterable}:
#     {statements}
# 
# Iterable: A collection of elements that can be iterated over (e.g., list, set, string, tuple, dictionary)
# statements: The block of code to be executed for each element in the iterable
# the number of iterations is determined by the number of elements in the iterable.
# The iterator variable takes the value of each element in the iterable one by one during each iteration.
numbers = [12,34,56,78,90]
for i in numbers:
    print("Hello")
    print(i)
```

# break / continue

```python
# break
# break statement is used to exit the loop prematurely when a certain condition is met.
for n in range(1,11):
    if n == 6:
        break
    print(n)
```

```python
#  continue
# continue statement is used to skip the current iteration and move to the next iteration of the loop.
for n in range(1,11):
    if n == 6 or n == 8:
        continue
    print(n)
```

# Function

```python
# Syntax
# def {function_name}({parameters}):
#     {statements}
# A function is a block of reusable code that performs a specific task.
#  Functions code shall only execute when the function is called.

def greet():
    print("Hello, welcome to the function world!")

greet()
greet()
```

```python
# Syntax
# def {function_name}({parameters}):
#     {statements}
# A function is a block of reusable code that performs a specific task.
#  Functions code shall only execute when the function is called.
#  Functions can take inputs (parameters) and can return outputs (return values).

def greet(name):
    print("Hello,", name, "! Welcome to the function world!")

greet("Islam")
greet("Ahmed")
```

```python
nums = [12,34,455,234,54]
def sum_list(numbers):
    summation = 0
    for n in numbers:
        summation += n
    print(summation)

sum_list(nums)
```

Quiz: Create a function that takes a list of numbers and returns the smallest number in the list.

```python
def find_smallest(numbers):
    smallest = numbers[0]
    for n in numbers[1:]:
        if n < smallest:
            smallest = n
    print(smallest)

nums = [12,34,455,234,5]
find_smallest(nums)
```
