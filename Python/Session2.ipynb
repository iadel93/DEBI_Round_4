# Basic Data Structure
- Sequential: List, Tuple
- Non-Sequential: Dictionary, set

```python
# List

data = [12,34,45,23,54,56, "Islam", 3.4]
print(data)
print(type(data))

# Access / Slicing
print(data[2])
print(data[5])
print(data[-1])

print(data[2:5])

# Modification
data[2] = 70
print(data)
```

```python
data = [12,43,64,4,56,20]

# Add items
data.append(20)
print(data)

data.insert(2, 50)
print(data)

tmp = [1,2,3]
# data.append(tmp) # [12, 43, 50, 64, 4, 56, 20, [1, 2, 3]]
data.extend(tmp)
print(data)

# Remove items
data.pop()
print(data)

data.pop(2)
print(data)

data.remove(56)
print(data)

data.remove(20) # Remove first occurrence only
print(data)
```

```python
data = [12,43,64,4,56,20]

# Sorting
data.sort()
print(data)

data.sort(reverse=True)
print(data)

# counting ,  index
print(data.count(20))
print(data.index(43))

# + Operator
data2 = [1,2,3]
data3 = data + data2
print(data3)

# Repetition
data4 = data2 * 3
print(data4)
```

```python
x = 3
y = x
y += 1
print(x)
print(y)
```

```python
# Copy
# Shallow Copy
data1 = [1,2,3]
data2 = data1  # Copy by reference
data2.append(4)

print(data1)
print(data2)
```

```python
# Deep Copy
data1=[1,2,3]
data2 = data1.copy() # Copy by values
data2.append(4)
print(data1)
print(data2)
```

# Dictionary

```python
#  Syntax:
# data = {'Key':'Value'}
# Key -> Can be int or str
# Value -> can be any type in python

data = {'Name':'Islam Adel', 'Id': 12345, 'Location':'Cairo'}

print(data)
print(type(data))

# Access
print(data['Name'])
print(data['Location'])

# Modify
data['Id'] = 102030
print(data)
```

```python
data = {'Name':'Islam Adel', 'Id': 12345, 'Location':'Cairo'}

# Add items
data['Grades'] = [10,20]
print(data)
data['Grades'] = [1,2]
print(data)

data.setdefault('Position', 'Lead')
print(data)
data.setdefault('Position', 'Senior')
print(data)

data2 = {'A':1, 'B':2}
data.update(data2)
print(data)

# Removing Items
data.pop('A')
print(data)

data.popitem()
print(data)

data.clear()
print(data)
```

```python
data = {'Name':'Islam Adel', 'Id': 12345, 'Location':'Cairo'}

print(data.items())
print(data.keys())
print(data.values())

# Copy 
data_copy = data.copy()
data_copy.popitem()
print(data)
print(data_copy)

# using get method
# print(data['Salary'])  # Prone to error if no such key
print(data.get('Name','Not Found'))
print(data.get('Salary','Not Found'))
```
